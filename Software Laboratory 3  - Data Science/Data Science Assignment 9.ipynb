```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

# Boxplot
The box represents the interquartile range (IQR), which is the middle 50% of the data, and the line inside the box represents the median age. The whiskers extend to the most extreme data points within 1.5 times the IQR, and any points beyond the whiskers are considered outliers.



```python

sns.boxplot(x='sex', y='age', hue='survived', data=titanic)
```

# Observations

1. The median age of both male and female passengers who survived is higher than the median age of those who did not survive.

 2.The age range of male passengers who did not survive is wider than the age range of male passengers who survived.
 
   
  3. The age range of female passengers who survived is wider than the age range of female passengers who did not survive.
   
  4. It seems that age was a factor in determining whether a passenger survived or not, with younger passengers having a higher chance of survival. However, this trend is more pronounced among male passengers, as the age range of male passengers who did not survive is wider than the age range of male passengers who survived. Among female passengers, age seems to have less of an impact on survival, as the age range of female passengers who survived is wider than the age range of female passengers who did not survive.
