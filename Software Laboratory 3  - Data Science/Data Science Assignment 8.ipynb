# Part 1

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

```

```python
titanic
```

```python
titanic.head(40)
```

```python
# plotting countplot to see how many passengers survived and did not survive
sns.countplot(x='survived', data=titanic)
```

```python
sns.catplot(data=titanic, x='age', y='pclass', kind='violin')
plt.show()
```

```python
x = 'Fare'
plt.hist(x, bins=5)
plt.show()
```

```python
g = sns.catplot(x="age",y="fare",
				hue="sex",
				data=titanic)
```

```python
# Catplot of kind swarm to see how passenger class, sex, and age relate to survival rate
sns.catplot(x='pclass', y='age', hue='sex', col='survived', kind='swarm', data=titanic)
```

# Catplot 
 Catplot describes that survival rate with respect to age, sex and pclass 
 
 Survived     : 1. Female people survived in more number than male
 
        2. pclass = 1 passenger survived in more number
                
 Not Survived : 1. Male people died in more number than female
 
        2. pclass= 3 passenger died in more number 
 
 
 
 
 
 

```python
# Heatmap to see how the fare was distributed among different classes
fare_by_class = titanic.pivot_table(index='class', values='fare', aggfunc='mean')
sns.heatmap(fare_by_class, cmap='YlOrRd', annot=True)
plt.title('Heatmap')
```

# Heatmap
1.Plotting of correlation between fare paid by passenger and survival rate.

2.Passengers who paid a higher fare got good facilities so had a higher survival rate .

#  Part 2  
Write a code to check how the price of the ticket (column name: 'fare') for each passenger is distributed by plotting a histogram.

```python
# Histogram showing price of ticket(fare) for each passenger 
sns.histplot(x='fare', data=titanic, kde=False)
plt.title("Histplot")
```

# Conclusion
Histogram shows a distribution of ticket prices, with the x-axis representing the price and the y-axis representing the count of passengers. Histogram shows that most of passenger fare was between 0 - 100 range , very few people afforded costly fare. Survival rate of pclass 1 was more due to good facilities . People with less age group survived just like children . Also ladies survived in more number than men.
