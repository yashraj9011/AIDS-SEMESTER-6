```python
import numpy as np
from matplotlib import pyplot as plt

# Sigmoidal
def tanh_function(z):
    return (np.exp(z) - np.exp(-z)) / (np.exp(z) + np.exp(-z))

# Sigmoidal derivative
def tanh_prime_function(z):
    return 1 - np.power(tanh_function(z),2)

# Generating data to plot
x_data = np.linspace(-6,6,100)
y_data = tanh_function(x_data)
dy_data = tanh_prime_function(x_data)
 
# Plotting
plt.plot(x_data, y_data, x_data,dy_data)
plt.title('Tanh Function & Derivative')
plt.legend(['f(x)','f\'(x)'])
plt.grid()
plt.show()
```

```python
def Softmax(x):
    '''
    Performs the softmax activation on a given set of inputs
    Input: x (N,k) ndarray (N: no. of samples, k: no. of nodes)
    Returns: 
    Note: Works for 2D arrays only(rows for samples, columns for nodes/outputs)
    '''
    max_x = np.amax(x, 1).reshape(x.shape[0],1) # Get the row-wise maximum
    e_x = np.exp(x - max_x ) # For stability
    return e_x / e_x.sum(axis=1, keepdims=True) 
```

```python
import numpy as np
from matplotlib import pyplot as plt

# Swish function
def sig(x):
    return x * 1/(1+np.exp(-x))

# Swish derivative  
def dsig(x):
    return  x * 1/(1+np.exp(-x)) + 1/(1+np.exp(-x)) * (1-x * 1/(1+np.exp(-x)))

# Generating data to plot
x_data = np.linspace(-8,8,50)
y_data = sig(x_data)
dy_data = dsig(x_data)

# Plotting
plt.plot(x_data, y_data,dy_data)
plt.title('Swish Function & Derivative')
plt.legend(['f(x)','f\'(x)'])
plt.grid()
plt.show()
```
